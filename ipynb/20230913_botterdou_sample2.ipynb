```python
import time
import numpy as np
import python_bitbankcc

# APIキーとシークレットを環境変数から取得
print("APIキーとシークレットを環境変数から取得します。")
API_KEY = 'BITBANK_API_KEY'
API_SECRET = 'BITBANK_API_SECRET'

# APIオブジェクトを作成
print("APIオブジェクトを作成します。")
pub = python_bitbankcc.public()
prv = python_bitbankcc.private(API_KEY, API_SECRET)

# 移動平均線の期間設定
print("移動平均線の期間を設定します。")
short_term = 1
long_term = 2

def calculate_moving_average(data, term):
    print(f"{term}期間の移動平均線を計算します。")
    return np.convolve(data, np.ones(term)/term, mode='valid')

def check_golden_cross(short_ma, long_ma):
    print("ゴールデンクロスをチェックします。")
    return short_ma[-1] > long_ma[-1] and short_ma[-2] <= long_ma[-2]

while True:
    # 現在の価格データを取得（例：5分足）
    print("現在の価格データを取得します。")
    value = pub.get_candlestick('btc_jpy', '1min', '20230830')
    ohlcv = value['candlestick'][0]['ohlcv']
    close_prices = [x[3] for x in ohlcv]
    print(close_prices)

    # 移動平均線を計算
    print("移動平均線を計算します。")
    short_ma = calculate_moving_average(np.array(close_prices, dtype=float), short_term)
    long_ma = calculate_moving_average(np.array(close_prices, dtype=float), long_term)
    print(short_ma)
    print(long_ma)

    # ゴールデンクロスをチェック
    if check_golden_cross(short_ma, long_ma):
        print("ゴールデンクロス発生！ 買い注文を出します。")

        # 買い注文（成行注文）
        print("成行で買い注文を出します。")
        result = prv.order(
            'btc_jpy',
            None,  # 価格（成行注文なのでNone）
            '0.0001',  # 注文量（BTC）
            'buy',  # 買い注文
            'market'  # 成行注文
        )
        print("注文結果:", result)
    else:
        print("ゴールデンクロスが発生していません。次のチェックまで待機します。")

    print("1分待機します。")
    time.sleep(60)  # 1分ごとにチェック
```
